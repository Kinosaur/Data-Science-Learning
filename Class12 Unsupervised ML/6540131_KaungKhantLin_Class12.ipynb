```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

plt.style.use('bmh')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.data)
```

```python
# Load the data from the CSV file
pointsDF = pd.read_csv('points.csv', index_col=0)
points = pointsDF.to_numpy()

# Extract xs and ys
xs = points[:, 0]
ys = points[:, 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(xs, ys, alpha=0.5)
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Scatter Plot of Points")
plt.show()
```

```python
# Load new points data
npointsDF = pd.read_csv('new_points.csv', index_col=0)
new_points = npointsDF.to_numpy()

# Create a KMeans instance with 3 clusters
model = KMeans(n_clusters=3, random_state=42)

# Fit model to points
model.fit(points)

# Determine the cluster labels of new_points
labels = model.predict(new_points)

# Print cluster labels of new_points
print('Labels: ', labels)
```

```python
# Assign the columns of new points: xs and ys
xs = new_points[:, 0]
ys = new_points[:, 1]

# Make a scatter plot of xs and ys, using labels to define the colors
plt.scatter(xs, ys, c=labels, alpha=0.5, cmap='viridis')

# Assign the cluster centers: centroids
centroids = model.cluster_centers_

# Assign the columns of centroids
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]

# Make a scatter plot of centroids_x and centroids_y
plt.scatter(centroids_x, centroids_y, marker='D', s=100, c='red', label='Centroids')

# Show the plot
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Clustered Data with Centroids")
plt.legend()
plt.show()
```

```python
# Load the data from the CSV file
seedsDF = pd.read_csv('seeds.csv')

# Convert DataFrame to NumPy array
seeds = seedsDF.to_numpy()

# Define range of k values
ks = range(1, 10)  # Changed to include 10 as well

# Initialize list to store inertia values
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters
    model = KMeans(n_clusters=k)
    
    # Fit model to seeds data
    model.fit(seeds)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)

# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')  # Capitalized label for readability
plt.ylabel('Inertia')
plt.xticks(ks)
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Create a KMeans instance with 3 clusters
model = KMeans(n_clusters=3)

# Fit model to samples
model.fit(seeds)

# Predict cluster labels
labels = model.predict(seeds)

# Extract xs and ys for the scatter plot
xs = seeds[:, 0]
ys = seeds[:, 6]

# Print shapes
print("Seeds shape:", seeds.shape)
print("Labels shape:", labels.shape)

# Make a scatter plot of xs and ys, using labels to define the colors
plt.scatter(xs, ys, c=labels, alpha=0.5)

# Assign the cluster centers
centroids = model.cluster_centers_
print("Centroids:\n", centroids)

# Assign the columns of centroids
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 6]

# Scatter plot of centroids
plt.scatter(centroids_x, centroids_y, marker='D', s=50, c='red', label='Centroids')

# Show the plot
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

```python
# Load fish dataset
fish = pd.read_csv('fish.csv', header=None)

# Extract samples (dropping the first column)
samples = fish.drop(0, axis=1) 

# Clustering without standardization
kmeans = KMeans(n_clusters=4)

# Fit KMeans to samples
kmeans.fit(samples)

# Calculate cluster labels
labels = kmeans.predict(samples)

# Create DataFrame with labels and species
df = pd.DataFrame({'labels': labels, 'species': species})

# Create cross-tabulation
ct = pd.crosstab(df['labels'], df['species'])

# Display cross-tabulation
print(ct)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Create a KMeans instance with 4 clusters
kmeans = KMeans(n_clusters=4)

# Create a pipeline with scaler and KMeans
pipeline = make_pipeline(scaler, kmeans)

# Fit the pipeline to the fish measurement samples
pipeline.fit(samples)

# Obtain cluster labels for samples
labels = pipeline.predict(samples)

# Create a DataFrame with labels and species
df = pd.DataFrame({'labels': labels, 'species': species})

# Create a cross-tabulation of cluster labels and species
ct = pd.crosstab(df['labels'], df['species'])

# Display cross-tabulation
print(ct)
```
